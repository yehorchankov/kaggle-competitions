```python
combined_df = combined_df.fillna(combined_df.median()[:'Fare'])
combined_df.loc[combined_df['Sex'] == 'male', 'Sex'] = 0
combined_df.loc[combined_df['Sex'] == 'female', 'Sex'] = 1
combined_df['Cabin'] = combined_df['Cabin'].fillna(0)
combined_df['NameLength'] = combined_df['Name'].apply(lambda x: len(x))
combined_df['IsChild'] = combined_df['Age'].apply(lambda x: 1 if x < 16 else 0)
combined_df['FamilySize'] = combined_df['SibSp'] + combined_df['Parch']
```

```python
import re
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, 
                 "Master": 4, "Dr": 5, "Rev": 6, 
                 "Major": 7, "Col": 7, "Mlle": 8, 
                 "Mme": 8, "Don": 9, "Lady": 10, 
                 "Countess": 10, "Jonkheer": 10, 
                 "Sir": 9, "Capt": 7, "Ms": 2}
regexp = '[\(\) ]{}\.'

def get_title(name):
    index = 0
    for k,v in title_mapping.items():
        if re.search(regexp.format(k), name):
            index = v
            break
    return index

titles = combined_df['Name'].apply(get_title)
combined_df['Titles'] = titles
```

```python
decks = ['A','B','C','D','E','F','G','T']

def get_deck(cabin):
    if cabin == 0:
        return 0
    
    for k,v in enumerate(decks):
        if cabin.startswith(v):
            return k+1
        
combined_df['Deck'] = combined_df['Cabin'].apply(get_deck)
```

```python
combined_export = combined_df[:891].filter(['Pclass', 'Survived', 'Sex', 'Age', 'SibSp', 'Parch'], axis=1)
combined_export['Survived'] = combined_export['Survived'].astype(int)
combined_export.to_csv("titanic3.csv", index=False)
```

```python
clear_train_df = combined_df[0:891].filter(['Survived','Cabin','Pclass','Sex','Age','Fare','FamilySize',
                                  'SibSp','Parch','NameLength','IsChild','Titles','Deck'], axis=1)
clear_train_df.info()
```

```python
clear_test_df = combined_df[891:].filter(['Survived','Cabin','Pclass','Sex','Age','Fare','FamilySize',
                                  'SibSp','Parch','NameLength','IsChild','Titles','Deck'], axis=1)
clear_test_df.info()
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
import matplotlib.pyplot as plt

predictors = ['Pclass', 'Sex', 'Age', 'Fare', 'FamilySize', 'Titles', 'SibSp', 'Parch', 'NameLength', 'Deck', 'IsChild']

selector = SelectKBest(f_classif, k=3)
selector.fit(clear_train_df[predictors], train_df['Survived'])

scores = -np.log10(selector.pvalues_)

plt.bar(range(len(predictors)), scores)
plt.xticks(range(len(predictors)), predictors, rotation='vertical')
plt.show()
```

# Implementing ML

## Random forest with KFold 

```python
from sklearn import cross_validation
from sklearn.ensemble import RandomForestClassifier

predictors = ['Pclass', 'Sex', 'Fare', 'Titles', 'NameLength', 'Deck']

forest = RandomForestClassifier(random_state=1, n_estimators=15, min_samples_split=8, min_samples_leaf=1)
kfold = cross_validation.KFold(clear_train_df.shape[0], n_folds=10, random_state=1)
scores = cross_validation.cross_val_score(forest, clear_train_df[predictors], clear_train_df['Survived'], cv=kfold)
print(scores.mean())
```

## Adaptive boosting with Decision Tree

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

tree = DecisionTreeClassifier(criterion='gini',max_depth=4)
adaboost = AdaBoostClassifier(base_estimator=tree, n_estimators=300, learning_rate=0.1,random_state=1)

adaboost = adaboost.fit(clear_train_df[predictors], clear_train_df['Survived'])
clear_train_pred = adaboost.predict(clear_train_df[predictors])
print(accuracy_score(clear_train_df['Survived'],clear_train_pred))
```

## Bagging classifier

```python
from sklearn.ensemble import BaggingClassifier

bag = BaggingClassifier(base_estimator=tree,
                        n_estimators=500, 
                        max_samples=1.0, 
                        max_features=1.0, 
                        bootstrap=True, 
                        bootstrap_features=False, 
                        n_jobs=4, 
                        random_state=1)

bag = bag.fit(clear_train_df[predictors], clear_train_df['Survived'])
clear_train_pred = bag.predict(clear_train_df[predictors])
print(accuracy_score(clear_train_df['Survived'],clear_train_pred))
```

## SVM

```python
from sklearn.svm import SVC

svc = SVC(C=1.0,
         kernel='rbf',
         random_state=1)

svc = svc.fit(clear_train_df[predictors], clear_train_df['Survived'])
clear_train_pred = svc.predict(clear_train_df[predictors])
print(accuracy_score(clear_train_df['Survived'],clear_train_pred))
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(tol=0.0001,
                             solver='liblinear',
                             random_state=1)

log_reg = log_reg.fit(clear_train_df[predictors], clear_train_df['Survived'])
clear_train_pred = log_reg.predict(clear_train_df[predictors])
print(accuracy_score(clear_train_df['Survived'],clear_train_pred))
```

## Voting classifier with Random Forest, Adaboost and Logistic Regression

```python
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import cross_val_score


voting_cls = VotingClassifier([('Random forest',forest),
                              ('Adaboost',adaboost),
                              ('Logistic Regression',log_reg)],
                             voting='hard')
cv_score = cross_val_score(voting_cls,
                          clear_train_df[predictors], 
                          clear_train_df['Survived'],
                          cv=10,
                          scoring='accuracy')
print(cv_score.mean())
```

```python
voting_cls = voting_cls.fit(clear_train_df[predictors], clear_train_df['Survived'])
```

# Predicting

```python
clear_test_pred = voting_cls.predict(clear_test_df[predictors])

submission = pd.DataFrame({
        "PassengerId": test_df["PassengerId"],
        "Survived": clear_test_pred.astype(int)
    })
submission.to_csv("kaggle.csv", index=False)
```
